```python
import pandas as pd
from pathlib import Path
import numpy as np
```

```python
# Pull data from S3 and create dataframe
url = 'https://airport-efficiency.s3.amazonaws.com/Merged_Raw_Flight_Data.csv'
flight_df = pd.read_csv(url)
```

```python
flight_df.head(5)
```

```python
# drop unnamed column that was imported from csv index
flight_df = flight_df.drop('Unnamed: 0', 1)
```

```python
# check data types
flight_df.dtypes
```

```python
#create seperate dataframes for each airport and reset index to start at 0
DCA_flights = flight_df.loc[flight_df['ORIGIN']== 'DCA']
DCA_flights = DCA_flights.reset_index(drop=True)
```

```python
DCA_flights['ORIGIN'].value_counts()
```

```python
DCA_flights['YEAR'].value_counts()
```

```python
IAD_flights = flight_df.loc[flight_df['ORIGIN']== 'IAD']
IAD_flights = IAD_flights.reset_index(drop=True)
```

```python
IAD_flights['ORIGIN'].value_counts()
```

```python
BWI_flights = flight_df.loc[flight_df['ORIGIN']== 'BWI']
BWI_flights = BWI_flights.reset_index(drop=True)
```

```python
BWI_flights['ORIGIN'].value_counts()
```

```python
# Fill missing State names
states = {"AL":"Alabama",
"AK":"Alaska",
"AZ":"Arizona",
"AR":"Arkansas",
"CA":"California",
"CO":"Colorado",
"CT":"Connecticut",
"DE":"Delaware",
"FL":"Florida","GA":
"Georgia",
"HI":"Hawaii",
"ID":"Idaho",
"IL":"Illinois",
"IN":"Indiana",
"IA":"Iowa",
"KS":"Kansas",
"KY":"Kentucky",
"LA":"Louisiana",
"ME":"Maine",
"MD":"Maryland",
"MA":"Massachusetts",
"MI":"Michigan",
"MN":"Minnesota",
"MS":"Mississippi",
"MO":"Missouri",
"MT":"Montana",
"NE":"Nebraska",
"NV":"Nevada",
"NH":"New Hampshire",
"NJ":"New Jersey",
"NM":"New Mexico",
"NY":"New York",
"NC":"North Carolina",
"ND":"North Dakota",
"OH":"Ohio",
"OK":"Oklahoma",
"OR":"Oregon",
"PR":"Puerto Rico",          
"PA":"Pennsylvania",
"RI":"Rhode Island",
"SC":"South Carolina",
"SD":"South Dakota",
"TN":"Tennessee",
"TX":"Texas",
"UT":"Utah",
"VI":"U.S. Virgin Islands",          
"VT":"Vermont",
"VA":"Virginia",
"WA":"Washington",
"WV":"West Virginia",
"WI":"Wisconsin",
"WY":"Wyoming"
}

```

```python
IAD_flights.head()
```

```python
# loop though dataframe and set state name based on abbreviation
for row in range(len(IAD_flights['DEST_STATE_ABR'])):
    abr = IAD_flights.at[row, 'DEST_STATE_ABR']
    if  abr in states.keys():
        IAD_flights.at[row ,'DEST_STATE_NM'] = states[abr]
IAD_flights = IAD_flights.drop('DEST_STATE_ABR', 1)
```

```python
IAD_flights.head(-10)
```

```python
IAD_flights['DEST_STATE_NM'].value_counts()
```

```python
for row in range(len(BWI_flights['DEST_STATE_ABR'])):
    abr = BWI_flights.at[row, 'DEST_STATE_ABR']
    if  abr in states.keys():
        BWI_flights.at[row ,'DEST_STATE_NM'] = states[abr]
BWI_flights = BWI_flights.drop('DEST_STATE_ABR', 1)
```

```python
for row in range(len(DCA_flights['DEST_STATE_ABR'])):
    abr = DCA_flights.at[row, 'DEST_STATE_ABR']
    if  abr in states.keys():
        DCA_flights.at[row ,'DEST_STATE_NM'] = states[abr]
DCA_flights = DCA_flights.drop('DEST_STATE_ABR', 1)

```

```python
# Find null values
for column in IAD_flights.columns:
    print(f"{column} has {IAD_flights[column].isnull().sum()}null values")
```

```python
# Drop null values in listed columns
IAD_flights = IAD_flights.dropna( how='any',
                    subset=['YEAR','MONTH','DAY_OF_WEEK','FL_DATE','BRANDED_CODE_SHARE','ORIGIN','DEST', 'DEST_STATE_NM','CRS_DEP_TIME','DEP_TIME', 'DEP_DELAY', 'DEP_DELAY_GROUP', 'DEP_TIME_BLK'  ])
```

```python
# Fill null values in all columns should only be delay reasons columns and reset index to 0
IAD_flights = IAD_flights.fillna(0)
IAD_flights = IAD_flights.reset_index(drop=True)
```

```python
# Find null values
for column in IAD_flights.columns:
    print(f"{column} has {IAD_flights[column].isnull().sum()} null values")
```

```python
BWI_flights = BWI_flights.dropna( how='any',
                    subset=['YEAR','MONTH','DAY_OF_WEEK','FL_DATE','BRANDED_CODE_SHARE','ORIGIN','DEST', 'DEST_STATE_NM','CRS_DEP_TIME','DEP_TIME', 'DEP_DELAY', 'DEP_DELAY_GROUP', 'DEP_TIME_BLK'  ])
BWI_flights = BWI_flights.fillna(0)
BWI_flights = BWI_flights.reset_index(drop=True)
```

```python
DCA_flights = DCA_flights.dropna( how='any',
                    subset=['YEAR','MONTH','DAY_OF_WEEK','FL_DATE','BRANDED_CODE_SHARE','ORIGIN','DEST', 'DEST_STATE_NM','CRS_DEP_TIME','DEP_TIME', 'DEP_DELAY', 'DEP_DELAY_GROUP', 'DEP_TIME_BLK'  ])
DCA_flights = DCA_flights.fillna(0)
DCA_flights = DCA_flights.reset_index(drop=True)
```

```python
IAD_flights.head()
```

```python
# create delayed column
# 1 = delayed 15 minutes or longer 
# 0 = not delayed
for row in range(len(IAD_flights['DEP_DELAY'])):
    delay = IAD_flights.at[row, 'DEP_DELAY']
    if  delay >= 15:
        IAD_flights.at[row ,'DELAYED'] = 1
    else:
        IAD_flights.at[row ,'DELAYED'] = 0
```

```python
IAD_flights.head(-10)
```

```python
for row in range(len(DCA_flights['DEP_DELAY'])):
    delay = DCA_flights.at[row, 'DEP_DELAY']
    if  delay >= 15:
        DCA_flights.at[row ,'DELAYED'] = 1
    else:
        DCA_flights.at[row ,'DELAYED'] = 0
```

```python
for row in range(len(BWI_flights['DEP_DELAY'])):
    delay = BWI_flights.at[row, 'DEP_DELAY']
    if  delay >= 15:
        BWI_flights.at[row ,'DELAYED'] = 1
    else:
        BWI_flights.at[row ,'DELAYED'] = 0
```

```python
# Convert FL_DATE column to datetime to remove time
IAD_flights['DATE']= pd.to_datetime(IAD_flights['FL_DATE'])
```

```python
# drop FL_DATE columns no longer needed
IAD_flights = IAD_flights.drop('FL_DATE', 1)
```

```python
IAD_flights.head()
```

```python
# Covid column covid start date set as 2020-3-15
# 1 = coivd was around
# 0 = before covid

for row in range(len(IAD_flights['DATE'])):
    date = IAD_flights.at[row, 'DATE']
    if  date > pd.to_datetime('2020-3-15'):
        IAD_flights.at[row ,'COVID'] = 1
    else:
        IAD_flights.at[row ,'COVID'] = 0
```

```python
IAD_flights.head(-10)
```

```python
# Convert FL_DATE column to datetime to remove time
BWI_flights['DATE']= pd.to_datetime(BWI_flights['FL_DATE'])
# drop FL_DATE columns no longer needed
BWI_flights = BWI_flights.drop('FL_DATE', 1)
```

```python
for row in range(len(BWI_flights['DATE'])):
    date = BWI_flights.at[row, 'DATE']
    if  date > pd.to_datetime('2020-3-15'):
        BWI_flights.at[row ,'COVID'] = 1
    else:
        BWI_flights.at[row ,'COVID'] = 0
```

```python
# Convert FL_DATE column to datetime to remove time
DCA_flights['DATE']= pd.to_datetime(DCA_flights['FL_DATE'])
# drop FL_DATE columns no longer needed
DCA_flights = DCA_flights.drop('FL_DATE', 1)
```

```python
for row in range(len(DCA_flights['DATE'])):
    date = DCA_flights.at[row, 'DATE']
    if  date > pd.to_datetime('2020-3-15'):
        DCA_flights.at[row ,'COVID'] = 1
    else:
        DCA_flights.at[row ,'COVID'] = 0
```

```python
# Delay reason column number 1-5 which is the most influential for the delay of that flight
# No delay = 0
# CARRIER_DELAY = 1
# WEATHER_DELAY = 2	
# NAS_DELAY = 3
# SECURITY_DELAY = 4	
# LATE_AIRCRAFT_DELAY = 5

for row in range(len(IAD_flights['CARRIER_DELAY'])):
    delay_reason = [0]
    delay_reason.append(IAD_flights.at[row, 'CARRIER_DELAY'])
    delay_reason.append(IAD_flights.at[row, 'WEATHER_DELAY'])
    delay_reason.append(IAD_flights.at[row, 'NAS_DELAY'])
    delay_reason.append(IAD_flights.at[row, 'SECURITY_DELAY'])
    delay_reason.append(IAD_flights.at[row, 'LATE_AIRCRAFT_DELAY'])
    n = np.array(delay_reason)
    IAD_flights.at[row, 'DELAY_REASON'] = np.argmax(n)
```

```python
for row in range(len(BWI_flights['CARRIER_DELAY'])):
    delay_reason = [0]
    delay_reason.append(BWI_flights.at[row, 'CARRIER_DELAY'])
    delay_reason.append(BWI_flights.at[row, 'WEATHER_DELAY'])
    delay_reason.append(BWI_flights.at[row, 'NAS_DELAY'])
    delay_reason.append(BWI_flights.at[row, 'SECURITY_DELAY'])
    delay_reason.append(BWI_flights.at[row, 'LATE_AIRCRAFT_DELAY'])
    n = np.array(delay_reason)
    BWI_flights.at[row, 'DELAY_REASON'] = np.argmax(n)
```

```python
for row in range(len(DCA_flights['CARRIER_DELAY'])):
    delay_reason = [0]
    delay_reason.append(DCA_flights.at[row, 'CARRIER_DELAY'])
    delay_reason.append(DCA_flights.at[row, 'WEATHER_DELAY'])
    delay_reason.append(DCA_flights.at[row, 'NAS_DELAY'])
    delay_reason.append(DCA_flights.at[row, 'SECURITY_DELAY'])
    delay_reason.append(DCA_flights.at[row, 'LATE_AIRCRAFT_DELAY'])
    n = np.array(delay_reason)
    DCA_flights.at[row, 'DELAY_REASON'] = np.argmax(n)
```
