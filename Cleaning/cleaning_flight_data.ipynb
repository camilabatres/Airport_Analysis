```python
import pandas as pd
from pathlib import Path
```

```python
# Pull data from S3 and create dataframe
url = 'https://airport-efficiency.s3.amazonaws.com/Flight_Data.csv'
flight_df = pd.read_csv(url)
```

```python
flight_df.head(5)
```

```python
#drop unnamed column that was imported
flight_df = flight_df.drop('Unnamed: 0', 1)
```

```python
flight_df.dtypes
```

```python
#create seperate dataframes for each airport and reset index to start at 0
DCA_flights = flight_df.loc[flight_df['ORIGIN']== 'DCA']
DCA_flights = DCA_flights.reset_index(drop=True)
```

```python
DCA_flights['ORIGIN'].value_counts()
```

```python
DCA_flights['YEAR'].value_counts()
```

```python
IAD_flights = flight_df.loc[flight_df['ORIGIN']== 'IAD']
IAD_flights = IAD_flights.reset_index(drop=True)
```

```python
IAD_flights['ORIGIN'].value_counts()
```

```python
BWI_flights = flight_df.loc[flight_df['ORIGIN']== 'BWI']
BWI_flights = BWI_flights.reset_index(drop=True)
```

```python
BWI_flights['ORIGIN'].value_counts()
```

```python
# loop though dataframe and set state name based on abbreviation
for row in range(len(IAD_flights['DEST_STATE_ABR'])):
    abr = IAD_flights.at[row, 'DEST_STATE_ABR']
    if  abr in states.keys():
        IAD_flights.at[row ,'DEST_STATE_NM'] = states[abr]
        # print(abr)
        # print(states[abr])
```

```python
for row in range(len(BWI_flights['DEST_STATE_ABR'])):
    abr = BWI_flights.at[row, 'DEST_STATE_ABR']
    if  abr in states.keys():
        BWI_flights.at[row ,'DEST_STATE_NM'] = states[abr]
```

```python
for row in range(len(DCA_flights['DEST_STATE_ABR'])):
    abr = DCA_flights.at[row, 'DEST_STATE_ABR']
    if  abr in states.keys():
        DCA_flights.at[row ,'DEST_STATE_NM'] = states[abr]
```
